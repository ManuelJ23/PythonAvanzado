## Ejercicio 3

```python
import math

def calcular_area_perimetro(radio):
  perimetro = 2 * math.pi * radio
  area = math.pi * radio**2
  return f"Perímetro: {round(perimetro, 2)}, Área: {round(area,2)}"
```

```python
print(calcular_area_perimetro(5))
```

## Ejercicio 4

```python
def login(username, password, intentos):
  if username == "usuario1" and password == "1234":
    return True, intentos
  else:
    intentos += 1
    return intentos

login("usuario1", 1234, 0)
login("Usuario1", "1234", 1)
login("usuario1", "1234", 2)
```
