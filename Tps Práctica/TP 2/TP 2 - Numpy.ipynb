## Ejercicio 2

```python
#Guardamos la ruta del archivo en una variable para su fácil manejo

rutaArchivo = "/content/drive/MyDrive/Colab Notebooks/2° Cuatri Python/Archivos varios/stock.txt"
```

```python
#Abrimos el archivo

archivo = open(rutaArchivo, "r")
```

```python
#Leemos el archivo e imprimimos su contenido

contenidoArchivo = archivo.read()
print(contenidoArchivo)
```

```python
#Separamos todas las líneas del archivo y guardamos el contenido de cada una como items de una lista

lineasDelArchivo = contenidoArchivo.splitlines()
print(lineasDelArchivo)
```

```python
#Convertimos esa lista en un Numpy Array

arrLineasArchivo = np.array(lineasDelArchivo)
print(arrLineasArchivo)
```

```python
#Reformamos la forma del array, aprovechando su simetría, generando 3 columnas

matrizReformada = np.reshape(arrLineasArchivo, (-1,3))
print(matrizReformada)
```

```python
#Separamos ese array (Que ya es una matriz) en 3 vectores que contengan un item específico cada uno

vectorArticulos = matrizReformada[:, 0]
vectorPrecios = matrizReformada[:, 1].astype(float)
vectorStock = matrizReformada[:, 2].astype(int)

print(vectorArticulos)
print(vectorPrecios)
print(vectorStock)
```

## Ejercicio 3

```python
def consultar():
  articulo = input("Ingrese el nombre de un producto: ")
  articulo = articulo.lower()
  if articulo in vectorArticulos:
    posicion = np.where(vectorArticulos == articulo)
    print(f"Artículo: {vectorArticulos[posicion]}")
    print(f"Precio: {vectorPrecios[posicion]}")
    print(f"Stock: {vectorStock[posicion]}")
  else:
    print("No se encuentra el artículo ingresado")
```

## Ejercicio 4

```python
def comprar(producto):
  posicion = np.where(vectorArticulos == producto)
  if vectorStock[posicion] >= 1:
    vectorStock[posicion] -= 1
    return True
  else:
    return False
```

```python
consultar()
comprar("microondas")
consultar()
```

## Ejercicio 5

```python
def agregar(producto, precio, cantidad):
  global vectorArticulos, vectorPrecios, vectorStock
  producto = producto.lower()
  if producto not in vectorArticulos:
    vectorArticulos = np.append(vectorArticulos, producto)
    vectorPrecios = np.append(vectorPrecios, precio)
    vectorStock = np.append(vectorStock, cantidad)
    return True
  else:
    print("El artículo ya existe.")
    return False
```

```python
def reponer(producto, cantidad):
  if producto in vectorArticulos:
    posicion = np.where(vectorArticulos == producto)
    vectorStock[posicion] += cantidad
    return True
  else:
    print("El artículo no existe.")
    return False
```

```python
agregar("horno", 75000, 10)
consultar()
reponer("horno", 5)
consultar()
comprar("horno")
consultar()
```

```python
#Opcional 1

def todoEnUno(numerosPorBillete, inicioRangoNumeros, finRangoNumeros, cantidadParticipantes):
  billeteGanador = np.random.randint(low= inicioRangoNumeros, high= finRangoNumeros, size=(1,numerosPorBillete))
  matrizParticipantes = np.random.randint(low= inicioRangoNumeros, high= finRangoNumeros, size=(cantidadParticipantes,numerosPorBillete))
  coincidencias = np.sum(billeteGanador == matrizParticipantes, axis=1)
  return billeteGanador, matrizParticipantes, coincidencias

pruebaLoteria = todoEnUno(10,0, 31, 5)
print(f"Billete ganador:\n {pruebaLoteria[0]}")
print(f"Billetes participantes:\n {pruebaLoteria[1]}")
print(f"Coincidencias:\n {pruebaLoteria[2]}")
```

## Ejercicio 7

```python
# Señal a decodificar
senal = '010010000110111101101100011000010100110101110101011011100110010001101111'
```

```python
# Convertir la señal en un array de enteros de 0 y 1
bits_array = np.array(list(senal), dtype=int)

# Reformar el array en un conjunto de filas de 8 columnas (paquetes de 8 bits)
bits_reshaped = bits_array.reshape(-1, 8)
bits_reshaped
```

```python
# Crear un vector de potencias de 2 para convertir de binario a decimal
potencias_dos = 2 ** np.arange(7, -1, -1)


# Multiplicar cada paquete de 8 bits por las potencias de 2 y sumar para obtener el valor decimal
valores_decimales = bits_reshaped.dot(potencias_dos)
valores_decimales
```

```python
# Convertir los valores decimales a caracteres ASCII
mensaje = ''.join(np.char.mod('%c', valores_decimales))

# Mostrar el mensaje decodificado
print("Mensaje descifrado:", mensaje)
```
