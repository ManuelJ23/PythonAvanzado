## Ejercicio 6

```python
df_ejerc_6 = pd.read_csv(path_folder + "/share_mujer_todos.csv")
df_puestos_mujeres = df_ejerc_6.copy()
df_puestos_mujeres
```

```python
df_puestos_mujeres.info()
```

### Inciso a

Forma que se me ocurrió a **mí**

```python
df_fecha_separada = df_puestos_mujeres.copy()
df_fecha_separada[["año", "mes", "dia"]] = df_puestos_mujeres["fecha"].str.split("-", expand=True)
df_fecha_separada
```

```python
top_años_ocup = round(df_fecha_separada.groupby("año")["share_mujer"].mean(), 4)
top_años_ocup.sort_values(ascending=False)
```

Forma eficiente con funciones de pandas **(DatetimeIndex)**

```python
df_ord_fechas = df_puestos_mujeres.copy()
df_ord_fechas['year'] = pd.DatetimeIndex(df_ord_fechas['fecha']).year
df_ord_fechas
```

```python
top_años_ocup = round(df_ord_fechas.groupby("year")["share_mujer"].mean(), 4)
top_años_ocup.sort_values(ascending=False)
```

El año **2023** fue el que tuvo más ocupación femenina, con un promedio de **40.69%**.

```python
df_puestos_mujeres.sort_values("share_mujer", ascending=False).head()
```

Siendo más específicos, en **Noviembre de 2023** tuvo lugar el mayor porcentaje de mujeres registradas ocupadas con un **45.10%** en la provincia de Chaco.

La provincia con mayor porcentaje de mujeres ocupadas a lo largo de los años es **La Pampa**, con un promedio de **42, 81%** de ocupación femenina en la provincia, seguida por **Jujuy** con un **42,00%**.
